Bibliotecas necessárias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Obtenção dos dados dos arquivos .csv

```python
meusDados = pd.read_csv('../arquivos/dadosLucas.csv')   
dadosModelo = pd.read_csv('../arquivos/dadosModelo.csv')
```

Manipulando os Dados

```python
# Fazendo a média das temperaturas
minhaMedia = meusDados['Temperatura'].mean()
mediaModelo = dadosModelo['Temperatura'].mean()

erroSist = mediaModelo - minhaMedia

print(f'Minha média = {minhaMedia:,.6f}')
print(f'Média modelo = {mediaModelo:,.6f}')
print(f'Erro sitemático = {erroSist:,.6f}')

tempAjust = meusDados['Temperatura'] + erroSist     # Temperatura ajustada
```

Plotagem:

```python
x = np.arange(0, 50, 5)    # As amostras foram coletadas com um intervalo de 5 segundos

plt.scatter(x, meusDados['Temperatura'])
plt.axhline(minhaMedia, color='purple', label='Média da Temperatura Medida (°C)')

plt.scatter(x, dadosModelo['Temperatura'], color='pink')
plt.axhline(mediaModelo, color='green',  label='Média da Temperatura Modelo (°C)', linewidth='4')

plt.axhline(minhaMedia + erroSist, color='red', label='Temperatura Ajustada (°C)', linestyle='--', linewidth='2')

plt.legend()
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (°C)')
plt.show()
```
